```python
selected_columns = ["short_name", "player_positions", "overall","club_name","league_name","nationality_name","preferred_foot","weak_foot"]  # Replace with actual column names
new_data = data[selected_columns]
```

```python
# preprocessing short_name column
```

```python
def preprocess_name(name):
    name = name.replace(".", "")  # Remove dots
    name = name.replace(" ", "_")  # Replace spaces with underscores
    name = name.lower()  # Convert to lowercase
    return name

new_data["short_name"] = new_data["short_name"].apply(preprocess_name)

```

```python
# preprocessing player_position column
```

```python
def preprocess_position(position):
    position = position.replace(",", " ")  # Replace commas with spaces
    position = position.lower()  # Convert to lowercase
    return position

# Apply the function to the 'position' column
new_data["player_positions"] = new_data["player_positions"].apply(preprocess_position)
```

```python
new_data.club_name
```

```python
# Function to preprocess club names
def preprocess_club_name(club_name):
    # Convert non-string values (like NaN or numbers) to an empty string
    if not isinstance(club_name, str):
        club_name = str(club_name) if club_name is not None else ""

    club_name = club_name.replace(".", "")  # Remove periods
    club_name = club_name.replace("-", " ")  # Replace hyphens with spaces
    club_name = " ".join(club_name.split())  # Remove extra spaces
    club_name = club_name.lower()  # Convert to lowercase
    club_name = club_name.replace(" ", "_")  # Replace spaces with underscores
    return club_name

# Apply the function to the 'club_name' column
new_data["club_name"] = new_data["club_name"].apply(preprocess_club_name)
```

```python
# preprocess league names same as club name
new_data["league_name"] = new_data["league_name"].apply(preprocess_club_name)
```

```python
# preprocess league names same as nationality
new_data["nationality_name"] = new_data["nationality_name"].apply(preprocess_club_name)
```

```python
# preprocess league names same as preferred_foot
def preprocess_club_name(preferred_foot):
    preferred_foot = preferred_foot.lower()  # Convert to lowercase
    return preferred_foot

# Apply the function to the 'club_name' column
new_data["preferred_foot"] = new_data["preferred_foot"].apply(preprocess_club_name)
```

```python
new_data = new_data.drop(columns=['weak_foot'])

print(new_data.head())
```

```python
# merging all columns into one:
new_data['description'] = new_data.apply(lambda row: ' '.join(row.astype(str)), axis=1)

# Display the updated DataFrame with the new 'description' column
print(new_data[['description']].head())
```

```python
description_data = new_data[['description']]
description_data
```

```python
import spacy

# Load spaCy's English tokenizer
nlp = spacy.load('en_core_web_sm')

# Tokenize the 'description' column in description_data
description_data['tokenized_description'] = description_data['description'].apply(lambda x: [token.text for token in nlp(x)])

# Display the updated DataFrame with the tokenized description

```

```python
description_data.tokenized_description
```

```python
model = gensim.models.Word2Vec(
    window=5,
    min_count=2
)
```

```python
model.build_vocab(description_data['tokenized_description'])
```

```python
model.train(description_data['tokenized_description'], total_examples=model.corpus_count, epochs=model.epochs)
```
